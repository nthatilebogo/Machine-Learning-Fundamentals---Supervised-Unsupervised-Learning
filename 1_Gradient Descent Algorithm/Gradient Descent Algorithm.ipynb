```python
import numpy as np

#This is the objective function
def obj_func(x):
        y = np.sin(x) + np.sin(10*x/3)
        return y

#This is the derivative function
def grad_func(x):
        y = np.cos(x) + (10/3) * np.cos(10*x/3)
        return y
```

```python
import matplotlib.pyplot as plt

#Search Bounds, lower and upper bound
xl = -1.3
xu = 0.5

x_n = np.arange(xl,xu,step=0.01)
y_n = obj_func(x_n)

plt.figure()
plt.plot(x_n,y_n,color='b')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Find the optimum point for f(x)=sin(x)+sin(10x/3)')
plt.show()
```

<h2>Gradient descent algorithm<h2>

```python
def gradDescAlgo(N_search_steps, x_i, gama):
    plt.figure()
    plt.plot(x_n,y_n,color='blue')
    for i in range(N_search_steps):
        #gradient descent search formula
        x_ip = x_i - gama*grad_func(x_i) 

        #gathering data points
        x_0 = x_i 
        x_1 = x_ip 

        #plot current step
        y_0 = obj_func(x_0)
        y_1 = obj_func(x_1)
        plt.scatter(x_0,y_0,color='red')
        plt.scatter(x_1,y_1,color='red')
        plt.plot([x_0,x_1],[y_0,y_1], color='green')

        #update x_i
        x_i=x_ip

    plt.title('Gradient descent search algorithm: Learning rate: %.2f' %(gama))
    plt.show()
    x_optimum = x_i
    y_opt = obj_func(x_optimum)
    print('optimum point: %.3f f-optimum:%.3f '%(x_optimum,y_opt))


N_search_steps = 100 #number of search steps
x_i = xl #initial value

gama = 0.2 #too large learning rate

gradDescAlgo(N_search_steps, x_i, gama)

gama = 0.01 #too small learning rate

gradDescAlgo(N_search_steps, x_i, gama)

gama = 0.05 #learning rate

gradDescAlgo(N_search_steps, x_i, gama)
```
